## 2. Prepraci√≥n y An√°lisis Exploratorio de Datos

````python
##  C√≥digo Preparaci√≥n de Datos.

import pandas as pd
import numpy as np

# leemos features
numerical_features = pd.read_parquet('train_numerical_features.parquet') #¬†features numericas
print(f"Dimensiones features num√©ricas: {numerical_features.shape}")
text_features = pd.read_parquet('train_text_features.parquet') # features de texto
text_features = text_features.drop(columns = ['title', 'tagline', 'credits']) # dropeamos columnas repetidas
print(f"Dimensiones features de texto: {text_features.shape}")

# mergeamos datasets
df = numerical_features.merge(text_features, on = 'id')
print(f'Dimensiones merge: {df.shape}')

df = df.drop(columns = ['poster_path', 'backdrop_path', 'recommendations']) # drop de columnas
df = df[df['revenue'] != 0] # filtrar ejemplos con revenue = 0
df[~(pd.isnull(df['release_date'])) & ~(pd.isnull(df['runtime']))] # drop si release_date y run_time son nulos
df['release_date'] = pd.to_datetime(df['release_date']) # pasamos release_date a datetime
df = df[df['status'] == 'Released'] # conservamos las filas donde status = Released

text_columns = list(df.select_dtypes('object').columns) # features de texto
df[text_columns] = df.select_dtypes('object').fillna('') # rellenamos columnas string con ''
df['label'] = pd.cut(df['vote_average'], bins = [0, 5, 6, 7, 8, 10], # generamos categorias
              labels = ['Negative', 'Mixed', 'Mostly Positive', 'Positive', 'Very Positive'])

df = df.drop(columns = ['vote_average', 'id']) # drop de vote_average y id
df = df.rename(columns = {'revenue': 'target'}) # rename revenue -> target
df
````

````python
def get_classfeatures(feature: str, n = 5, ngram_range = (1, 1)):
    
    '''
    Funci√≥n que retorna las top n categor√≠as en frecuencia por clase. 
    '''
    
    def custom_tokenizer(text):
        return text.split('-')

    bow = CountVectorizer(tokenizer = custom_tokenizer, ngram_range = ngram_range) # instanciamos bow
    data_bow = bow.fit_transform(df[feature]).toarray() # obtenemos dummies (bow) por feature
    data_bow = pd.DataFrame(data_bow, columns = bow.get_feature_names_out()) # generamos dataframe con bow

    label_dummies = pd.get_dummies(df['label'])
    labels = list(label_dummies.columns)
    ranking_features = (data_bow.T @ label_dummies).reset_index()
    ranking_features.columns = [feature] + labels
    output = {label: list(ranking_features.sort_values(by = label, ascending = False)[feature][:n]) for label in labels}
    
    return output

class_credits = get_classfeatures(feature = 'credits', n = 10)
class_credits['Negative']
````

````python
from sklearn.feature_extraction.text import CountVectorizer

def get_topfeatures(feature: str, percent_features: float = .2, n_movies: int = 50, ngram_range = (1, 1)) -> list:
    
    '''
    Funci√≥n que recibe una columna con texto y categor√≠as dentro, 
    retorna una lista con el top (percent_features)% de las categor√≠as seg√∫n target
    feature: nombre de la columna a procesar (str)
    percent_features: top % a ser considerado (float)
    n_movies: n peliculas con mejor target (int)
    '''
    
    # especificamos errores
    if (percent_features > 1)  | (percent_features < 0):
        raise ValueError('percent_features debe estar entre 0 y 1, pelmazo!')
    if n_movies > len(df):
        raise ValueError('n_movies no puede ser mayor al largo del dataset, pelmazo!')
    
    # analyzer para separar creditos
    def custom_tokenizer(text):
        return text.split('-')
    
    # obtenemos 50 peliculas con mayor 
    top50_movies = df.sort_values(by = 'target', ascending = False).head(n_movies)

    bow = CountVectorizer(tokenizer = custom_tokenizer, ngram_range = ngram_range) # instanciamos bow
    data_bow = bow.fit_transform(top50_movies[feature]).toarray() # obtenemos dummies (bow) por feature
    data_bow = pd.DataFrame(data_bow, columns = bow.get_feature_names_out()) # generamos dataframe con bow
    
    total_features = data_bow.shape[1] # cantidad de categor√≠as
    n_features = int(total_features * percent_features) # obtenemos n categorias con percent_features
    
    top50_target = np.array(top50_movies['target'], ndmin = 2) # pasamos a array
    ranking_features = (top50_target @ data_bow).T.reset_index() # sumamos target por feature
    ranking_features.columns = [feature, 'target'] # cambiamos nombre columnas
    ranking_features = ranking_features.sort_values(by = 'target', ascending = False) # rankeamos por target
    
    # devolvemos lista de top features con mejor target
    return list(ranking_features[feature])[:n_features]

top_credits = get_topfeatures(feature = 'credits', percent_features = .1, ngram_range = (1,2))
top_genres = get_topfeatures(feature = 'genres', percent_features = .1, ngram_range = (1,2))
top_companies = get_topfeatures(feature = 'production_companies', percent_features = .1, ngram_range = (1,2))
````

````python
from itertools import combinations, permutations

def make_dummies(feature: str, top_features: list, ngram_range = 1):
    
    '''
    Funci√≥n que genera dummies, 1 si el token esta contenido en lista top_features, 0 sino.
    '''
    
    def tokenizer(text: str, ngram_range = 1):
        text = text.split('-') #¬†tokenizamos
        text = [token.lower() for token in text] # pasamos a min√∫sculas

        # si usamos ngramas, sumamos a los tokens
        if ngram_range > 1:
            # obtenemos todas las permutaciones, los pares pueden estar desordenados!
            ngrams = list(permutations(text, ngram_range)) 
            ngrams = [' '.join(ngram) for ngram in ngrams] # join de pares

            text = text + ngrams # sumamos a lista de tokens

        return text
    
    df_tmp = df.copy() # generamos copia de dataframe
    df_tmp[f'{feature}_dummy'] = df_tmp[feature].apply(lambda x: tokenizer(x, ngram_range = ngram_range)) # separamos credits
    
    # 1 si pelicula contiene algun topcredit, 0 en caso contrario
    return df_tmp[f'{feature}_dummy'].apply(lambda x: 1 if len(set(x) & set(top_features)) > 0 else 0) 

df['credits_dummy'] = make_dummies(feature = 'credits', top_features = top_credits, ngram_range = 2)
df['genres_dummy'] = make_dummies(feature = 'genres', top_features = top_credits, ngram_range = 2)
df['companies_dummy'] = make_dummies(feature = 'production_companies', top_features = top_credits, ngram_range = 2)
df['year'] = df['release_date'].apply(lambda x: x.year).astype('category')
df['month'] = df['release_date'].apply(lambda x: x.month).astype('category')
df['classnegative_dummy'] = make_dummies(feature = 'credits', top_features = class_credits['Negative'])
df['classmixed_dummy'] = make_dummies(feature = 'credits', top_features = class_credits['Mixed'])
df['classmpositive_dummy'] = make_dummies(feature = 'credits', top_features = class_credits['Mostly Positive'])
df['classpositive_dummy'] = make_dummies(feature = 'credits', top_features = class_credits['Positive'])
df['classvpositive_dummy'] = make_dummies(feature = 'credits', top_features = class_credits['Very Positive'])
df
````

### Anexo: Generaci√≥n de Archivo Submit de la Competencia

Para subir los resultados obtenidos a la pagina de CodaLab utilice la funci√≥n `generateFiles` entregada mas abajo. Esto es debido a que usted deber√° generar archivos que respeten extrictamente el formato de CodaLab, de lo contario los resultados no se veran reflejados en la pagina de la competencia.

Para los resultados obtenidos en su modelo de clasificaci√≥n y regresi√≥n, estos ser√°n guardados en un archivo zip que contenga los archivos `predicctions_clf.txt` para la clasificaci√≥n y `predicctions_rgr.clf` para la regresi√≥n. Los resultados, como se comento antes, deberan ser obtenidos en base al dataset `test.pickle` y en cada una de las lineas deberan presentar las predicciones realizadas.

Ejemplos de archivos:

- [ ] `predicctions_clf.txt`

        Mostly Positive
        Mostly Positive
        Negative
        Positive
        Negative
        Positive
        ...

- [ ] `predicctions_rgr.txt`

        16103.58
        16103.58
        16041.89
        9328.62
        107976.03
        194374.08
        ...



````python
from zipfile import ZipFile
import os

def generateFiles(predict_data, clf_pipe, rgr_pipe):
    """Genera los archivos a subir en CodaLab

    Input
    predict_data: Dataframe con los datos de entrada a predecir
    clf_pipe: pipeline del clf
    rgr_pipe: pipeline del rgr

    Ouput
    archivo de txt
    """
    y_pred_clf = clf_pipe.predict(predict_data)
    y_pred_rgr = rgr_pipe.predict(predict_data)
    
    with open('./predictions_clf.txt', 'w') as f:
        for item in y_pred_clf:
            f.write("%s\n" % item)

    with open('./predictions_rgr.txt', 'w') as f:
        for item in y_pred_rgr:
            f.write("%s\n" % item)

    with ZipFile('predictions.zip', 'w') as zipObj2:
       zipObj2.write('predictions_rgr.txt')
       zipObj2.write('predictions_clf.txt')

    os.remove("predictions_rgr.txt")
    os.remove("predictions_clf.txt")
````
